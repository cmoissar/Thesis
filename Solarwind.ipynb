## Fourier transform

### Definition

```python
def plot_PSD_via_Fourier(n, signal, tmax):
    
    dt = tmax / n

    if n%2 == 0:
        nyquist = int(n/2)
    if n%2 == 1:
        nyquist = int((n-1)/2)     

    f = np.fft.fftfreq(n, d=dt)
    A = np.fft.fft(signal,n) / n
    f = f[:nyquist]
    A = A[:nyquist]

    PSD = 4*np.abs(A)**2

    plt.close('all')

    plt.figure(0)
    plt.plot(t,signal)
    plt.figure(1)
    plt.plot(f,PSD)
    plt.xlim([0, 8])

    plt.show()

    return
```

### Signal with constant spectral content

```python
n = 100000
tmax = 10
t = np.arange(0,n) * tmax / n
```

```python
a = 2*np.sin((2*pi)*t)
b = 4*np.sin(5*(2*pi)*t)

signal = a + b

plot_PSD_via_Fourier(n ,signal, tmax)
```

### Signal with varying spectral content

```python
a = 2*np.sin((2*pi)*t)
b = 4*np.sin(5*(2*pi)*t)

signal = np.zeros(n)

signal[:int(n/2)] = a[:int(n/2)]
signal[int(n/2)+1:] = b[int(n/2)+1:]
    
plot_PSD_via_Fourier(n , signal, tmax)
```
